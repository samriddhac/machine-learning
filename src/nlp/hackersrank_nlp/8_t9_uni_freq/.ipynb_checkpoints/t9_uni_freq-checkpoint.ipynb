```python
char_dict = {
    2:['a','b','c'],
    3:['d','e','f'],
    4:['g','h','i'],
    5:['j','k','l'],
    6:['m','n','o'],
    7:['p','q','r','s'],
    8:['t','u','v'],
    9:['w','x','y','z']
}
```

```python
def word_seq(text):
    seq = []
    text_len = len(text)
    num_item = text[0]
    if int(num_item) in char_dict.keys():
        char_seq = char_dict[int(num_item)]
        for c in char_seq:
            in_c = c
            in_word = text[1:]
            word_sequences(in_word, in_c, seq)
    
    return seq
    
    
def word_sequences(text, in_c, seq):
    out_c = ''
    if len(text)>0:
        num_item = text[0]
        char_seq = char_dict[int(num_item)]
        for c in char_seq:
            out_c = in_c + c
            if len(text) > 0:
                in_word = text[1:]
                word_sequences(in_word, out_c, seq)
    else:
        seq.append(in_c)
```

```python
texts = []
num = input()
for i in range(int(num)):
    text = input()
    texts.append(text.strip())
```

```python
def word_tokenize(texts):
    words = []
    sentence_punct_splitter = ['.', '?']
    sents = texts.split('\n')
    for char in sentence_punct_splitter:
        s_array = []
        for s in sents:
            s_split = s.split(char)
            s_array = s_array + [s_item.strip().lower() for s_item in s_split]
        sents = s_array
    for sentence in sents:
        w = sentence.split()
        words = words + w
    return words
```

```python
from collections import Counter
with open('t9Dictionary', 'r', encoding='utf-8') as f:
    text = f.read()
    dictionary_words = text.lower().split('\n')
print(dictionary_words[:20])
```

```python
corpus_word_counts = Counter()
with open('t9TextCorpus', 'r', encoding='utf-8') as fp:
    for line in fp:
        corpus_word_counts.update(line.split())\
```

```python
corpus_dict = dict(corpus_word_counts)
```

```python
dictionary_counter = Counter(dictionary_words)
```

```python
dictionary_dict = dict(dictionary_counter)
```

```python
word_count_dict = {}
for dict_key in dictionary_dict.keys():
    if dict_key in corpus_dict.keys():
        word_count_dict[dict_key] = int(dictionary_dict[dict_key])+int(corpus_dict[dict_key])
    else:
        word_count_dict[dict_key] = int(dictionary_dict[dict_key])
```

```python
import operator

for text in texts:
    w_seq = word_seq(text)
    search_res = {}
    for w in w_seq:
        items = {key:value for key, value in word_count_dict.items() if key.startswith(w) }
        for key, value in items.items():
            search_res[key] = value
    sorted_search_res = sorted(search_res.items(), key=operator.itemgetter(1), reverse=True)
    result_list = [key for key,value in sorted_search_res]
    if result_list is not None and len(result_list)>0:
        print(";".join(result_list[:5]))
    else:
        print('No Suggestions')
        
        
```
